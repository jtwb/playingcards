```python
import math

def riffle(n):
    ''' generate riffle shuffle permutation for deck of size n '''
    if n % 2:
        raise ValueError('n must be even')
        
    perm = []
    half = int(n/2)
    for i in xrange(half):
        perm.append(half+i)
        perm.append(i)

    return perm
```

```python
def applyperm(permutation):
    nnext = permutation
    while True:
        yield nnext
        nnext = [nnext[i] for i in permutation]
```
